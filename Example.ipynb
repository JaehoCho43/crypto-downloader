```python
from downloader import *
```

```python
# Initialize the downloader

d = Downloader(
    time_start = '2021-01-01', 
    time_end = '2022-05-01', 
    period = 'h'
)
```

```python
## Clear the directory

d.reset_data()
```

```python
## Get data for given tickers
## Downloaded data will be stored in the default directory 'data'

df = d.get_data(['BTC','ETH','XRP'])
```

```python
df
```

```python
df['close']
```

```python
## If we call get_data function with updated tickers, it will download the data only for new tickers
## To update the data for pre-existing ticker, remove the original file and run again

df = d.get_data(['BTC','ETH','XRP','BNB'])
```
